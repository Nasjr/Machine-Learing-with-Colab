Importing libraries 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#For data splitting 
from sklearn.model_selection import train_test_split
#Logistic Regression import
from sklearn.linear_model import LogisticRegression
#for accuracy
from sklearn.metrics import accuracy_score
```

Data collection and Data Processing 




```python
#loding data set
sonar=pd.read_csv('/content/Sonar Data.csv',header=None)
```

```python
sonar.head()
```

```python
#Number of Rows and Columns
sonar.shape
```

```python
#No Missing Data 
sonar.info()
```

```python
sonar.describe()
```

```python
sonar[60].value_counts()
#There is 111 Mine And 97 Rock prediction will be Good because the numbers are close
```

```python
sonar.groupby(60).mean()
```

```python
#Seprating data and labels X will contain all values except the Final one Y contians the Final value
x=sonar.drop(columns=60,axis=1)
y=sonar[60]
```

**Training and Test Data**

```python
#Stratify splits the data into almost equal no. of R or M in train data , Random state to reproduce the code as it is
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,stratify=y,random_state=1)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

**Model Traingin using Logistic Regressiong**

```python
model=LogisticRegression()
```

```python
#training the model into logistic Reg model with Training data
model.fit(x_train,y_train)
```

**Evaluate The Model**

```python
#accuracy on training data above 70% is Good
x_train_pred=model.predict(x_train)
training_data_acc=accuracy_score(x_train_pred,y_train)
print('Accuracy on training data :',training_data_acc)
```

**The Accuracy is 80% Because it is on the Training Data**

```python
x_test_pred=model.predict(x_test)
Test_data_accuracy=accuracy_score(x_test_pred,y_test)
print('Accuracy on Test data :',Test_data_accuracy)
```

The accuracy is 76% which is Good with respect to the size of the data
